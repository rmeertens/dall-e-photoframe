```python
import subprocess
import random
import os

import base64
from io import BytesIO

import requests
from PIL import Image, ImageDraw, ImageFont

class ServiceError(Exception):
    def __init__(self, status_code):
        self.status_code = status_code


def get_images_from_backend(prompt, backend_url):
    r = requests.post(backend_url, json={"prompt": prompt})
    if r.status_code == 200:
        json = r.json()
        images = json["images"]
        images = [Image.open(BytesIO(base64.b64decode(img))) for img in images]
        version = json.get("version", "unknown")
        return {"images": images, "version": version}
    else:
        raise ServiceError(r.status_code)


def get_model_version(url):
    r = requests.get(url)
    if r.status_code == 200:
        version = r.json()["version"]
        return version
    else:
        raise ServiceError(r.status_code)
        
```

```python

def add_frame_to_image(image: Image, description, frame_parameters): 
    frame_image = Image.open(frame_parameters['path'])
    
    # resize the pillow image to fit the frame size
    image = image.resize(frame_parameters['center_size'])

    # Create a white RGBA image of the same size as the frame
    output = Image.new('RGBA', frame_image.size, (255,255,255,0))

    # paste the image in the frame
    output.paste(image, frame_parameters['center_offset'])
    output.paste(frame_image, (0,0), frame_image)
    
    frame_image = output.resize(frame_parameters['out_size'])

    # Add the description using Pillow
    draw = ImageDraw.Draw(frame_image)
    font = ImageFont.truetype("/Library/Fonts/Arial.ttf", frame_parameters['font_size'])
    w, h = draw.textsize(description, font=font)

    top_left = (frame_parameters['text_center'][0] - w/2, frame_parameters['text_center'][1] - h/2)
    EDGE_PADDING = 10
    draw.rectangle([top_left[0]- EDGE_PADDING, top_left[1], top_left[0]+w+EDGE_PADDING, top_left[1]+h], fill=(255,255,255,255))
    
    draw.text(top_left, description, font=font, fill=(0,0,0))
    
    return frame_image




# # NOTE: frames downloaded from https://freesvg.org/decorative-divider-240-frame-2
# FRAME_FOLDER = 'frames'
# # show each frame
# for frame_name in sorted(os.listdir(FRAME_FOLDER)):
#     if frame_name.endswith('.png'):
#         frame_path = os.path.join(FRAME_FOLDER, frame_name)
                
#         frame_parameters = {
#             'path': frame_path,
#             'center_size': (1700,1700),
#             'center_offset': (340, 340),
#             'out_size': (800,800),
#             'text_center': (400,730),
#             'font_size': 25
#         }

#         dall_e_image = Image.open('out_images/89405057.jpg')

#         # output = add_frame_to_image(dall_e_image, 'A nice image of something here', frame_parameters)

#         # show the ouput
#         # output.show()
```
